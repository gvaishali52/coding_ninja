Problem Statement :
- Find out which cities are generally chosen for IPL Matches.
- Find the top 10 cities where IPL matches have been conducted.
- Plot a bar chart to visualise it.
- Print the names of the city along with the number of matches conducted there.


Notes:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
IPL= pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
IPL['city'].value_counts()
```

```python
IPL.isnull().sum()
```

```python
IPL.fillna
```

```python
import seaborn as sns
sns.countplot(x='city', data=IPL)
plt.show()
```

```python
import seaborn as sns
IPL_top_10 = IPL.city.value_counts().nlargest(10)
plt.xticks(rotation=30)
sns.countplot(x='city', data=IPL, order=IPL_top_10.index)
plt.show()
```
